```python
from nltk.stem import WordNetLemmatizer
import string
from nltk.corpus import stopwords
```

```python
def textnormaliation(text):
    text=text.lower()
    spl_char=re.sub(r'[^a-zA-z]',' ',text)
    
    tokens=nltk.word_tokenize(spl_char)
    
    char=[word for word in tokens if word not in string.punctuation]
    #chars=[word for word in char if word not in set(stopwords.words('english'))]
    lr=WordNetLemmatizer()
    tag_list=pos_tag(char,tagset=None)
    
    lemma_words=[]
    for a,b in tag_list:
        if b.startswith('V'):  # Verb
            pos_val='v'
        elif b.startswith('J'): # Adjective
            pos_val='a'
        elif b.startswith('R'): # Adverb
            pos_val='r'
        else:
            pos_val='n'
        
        lemma_tokens=lr.lemmatize(a,pos_val)
        lemma_words.append(lemma_tokens)
    x=' '.join(lemma_words)
    return x
```

```python
def ask(text):
    split=text.split()
    words=[word for word in split if word not in set(stopwords.words('english'))]
    words=[word for word in words if word not in string.punctuation]
    lr=nltk.WordNetLemmatizer()
    lemma_words=[]
    tag_list=pos_tag(words)
    for a,b in tag_list:
        if b.startswith('V'):  #verb
            pos_val='v'
        elif b.startswith('J'): # Adjective
            pos_val='a'
        elif b.startswith('R'): # Adverb
            pos_val='r'
        else:
            pos_val='n'
        lemma_tokens=lr.lemmatize(a,pos_val)
        lemma_words.append(lemma_tokens)
    
    words=' '.join(lemma_words)
    
    question_lemma=textnormaliation(words)
    question_bow=cv.transform([question_lemma]).toarray()
    cosine_value=cosine_similarity(dataframe,question_bow)
    index_value=cosine_value.argmax()
    return data['Text Response'].loc[index_value]
```
