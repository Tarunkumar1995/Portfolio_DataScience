```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
```

```python
state=[]
confirmed_cases=[]
active_cases=[]
recoverd_cases=[]
deaths=[]
url=requests.get('https://www.mygov.in/covid-19')
soup=BeautifulSoup(url.text)
data=soup.find_all('div',class_='marquee_data view-content')
for i in data[0].find_all('span',class_='st_name'):    #Names
    if len(i)==0: 
        state.append(np.nan)
    else:
        state.append(i.text)
for i in data[0].find_all('span',class_='st_number'):   #confirmed_cases
    if len(i)==0: 
        confirmed_cases.append(np.nan)
    else:
        confirmed_cases.append(i.text)
        
for i in data[0].find_all('div',class_='tick-active'):   #active_cases
    if len(i)==0:
        active_cases.append(np.nan)
    else:
        active_cases.append(re.sub(r'[^0-9]','',str(i)))      
        
for i in data[0].find_all('div',class_='tick-discharged'):               #recovered cases
    if len(i)==0:
        recoverd_cases.append(np.nan)
    else:
        recoverd_cases.append(re.sub(r'[^0-9]','',str(i)))  
        
for i in data[0].find_all('div',class_='tick-death'):     #deaths
    if len(i)==0:
        deaths.append(np.nan)
    else:
        deaths.append(re.sub(r'[^0-9]','',str(i)))  
        

                                                                            
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style='darkgrid')

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.DataFrame()
df['State']=state
df['Confirmed_cases']=confirmed_cases
df['Active_cases']=active_cases
df['Recoverd_cases']=recoverd_cases
df['Deaths']=deaths
df
```

```python
df.dtypes
```

```python
def remove(x):
    x=x.apply(lambda x: re.sub(r'[^0-9]','',x))
    return x
for i in df.iloc[:,1:].keys():
    df[i]=remove(df[i])
```

```python
df['Confirmed_cases']=pd.to_numeric(df['Confirmed_cases'])
df['Active_cases']=pd.to_numeric(df['Active_cases'])
df['Recoverd_cases']=pd.to_numeric(df['Recoverd_cases'])
df['Deaths']=pd.to_numeric(df['Deaths'])
```
